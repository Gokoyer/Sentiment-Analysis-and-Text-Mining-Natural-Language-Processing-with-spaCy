```python
# Lab 1 ‚Äì BMI Analysis (Figure-2 Output, Windows Safe)
with open("hw.txt", "r") as file:
    values = [float(line.strip()) for line in file if line.strip()]

underweight = []
normal = []
overweight = []
obesity = []

person = 1
all_bmis = []

for i in range(0, len(values), 2):
    height_cm = values[i]
    weight_lb = values[i + 1]

    height_m = height_cm / 100
    weight_kg = weight_lb / 2.2
    bmi = weight_kg / (height_m ** 2)

    all_bmis.append(bmi)

    if bmi < 18.5:
        underweight.append((person, bmi))
    elif bmi < 25:
        normal.append((person, bmi))
    elif bmi < 30:
        overweight.append((person, bmi))
    else:
        obesity.append((person, bmi))

    person += 1


def print_group(title, group):
    print(f"{title}:")
    for p, bmi in group:
        print(f"Person- {p} : {bmi:.4f}")
    print(f"Group Size: {len(group)}")

    if group:
        avg = sum(bmi for _, bmi in group) / len(group)
        print(f"Average BMI for this group: {avg:.4f}")
    print()


# Print output (same order as Figure-2)
print_group("Under Weight", underweight)
print_group("Normal", normal)
print_group("Over Weight", overweight)
print_group("Obesity", obesity)

print(f"The file has records for {len(all_bmis)} persons")
print(f"Average BMI for all these {len(all_bmis)} persons: {sum(all_bmis)/len(all_bmis):.4f}")

#---------------------------------------------------------------------------------------------------#
#Part B
#Complete the functions related to the following:
#--------------------1.String Types---------------------------------#
#(a) simple string
simple_string = 'Hello!' + " I'm a very simple string"
print(simple_string)

#(b) multi-line string
#Note that the \n (newline) escape character automatically created
multi_line_string = """Hello I'm
a multi-line
string!"""
multi_line_string
print(multi_line_string)

#c normal string with escape sequences leading to a wrong file path
escaped_string = "hw.txt"
print(escaped_string)

#d raw string keeps the backlashes in its normal form
raw_string = r'hw.txt'
print(raw_string)

#e unicode string literals
string_with_unicode = 'H\u00e8llo!'
print(string_with_unicode)

#----------------------------2 String Operations------------------------------#

#Different ways of String concatenation
'Hello üòä' + ' and welcome ' + 'to Python üêç!'

'Hello üòä' ' and welcome ' 'to Python üêç!'

# concatenation of variables and literals
s1 = 'Python üíª!'
'Hello üòä ' + s1

# some more ways of concatenating strings
s2 = '--üêçPythonüêç--'
s2 * 5
#---------------------------------------------#
s1 + s2

# concatenating several strings together in parentheses
s3 = ('This '
      'is another way '
      'to concatenate '
      'several strings!')
s3

#-------------------------3 String indexing and slicing----------------------#

# creating a string
s = 'PYTHON'
s, type(s)

#String Indexing
# depicting string indexes
for index, character in enumerate(s):
    print('Character ->', character, 'has index->', index)
s[0], s[1], s[2], s[3], s[4], s[5]
s[-1], s[-2], s[-3], s[-4], s[-5], s[-6]

#string slicing
s = 'PYTHON'
#s, type(s)
s[:]
s[1:4]
s[:3], s[3:]
s[-3:]
s[:3] + s[3:]
s[:3] + s[-3:]

#--------------------Useful String Methods------------------------------#
#Case Conversions
s = 'python is great'
s.capitalize()
s.upper()
s.title()

#String Replace
s.replace('python', 'NLP')

#Numeric checks
'12345'.isdecimal()

'apollo11'.isdecimal()

#Alphabet checks
'python'.isalpha()

'number1'.isalpha()

#Alphanumeric checks
'total'.isalnum()

'abc123'.isalnum()

'1+1'.isalnum()

#String splitting and joining
s = 'I,am,a,comma,separated,string'
s.split(',')

' '.join(s.split(','))

# stripping whitespace characters
s = '   I am surrounded by spaces    '
s

s.strip()

sentences = 'Python is great. NLP is also good.'
print(sentences.split('.'))

sentences = 'Python is great. NLP is also good.'
print('\n'.join(sentences.split('.')))

sentences = 'Python is great. NLP is also good.'
print('\n'.join([sentence.strip()
                     for sentence in sentences.split('.')
                         if sentence]))

#--------------------------------------String formatting----------------------------#

#Simple string formatting expressions - old style
s = 'Hello %s' %('Python!')
print(s)

s = 'Hello %s %s' %('World!', 'How are you?')
print(s)

#Formatting expressions with different data types - old style
s = 'We have %d %s containing %.2f gallons of %s' %(2, 'bottles', 2.5, 'milk')
print(s)

s = 'We have %d %s containing %.2f gallons of %s' %(5.21, 'jugs', 10.86763, 'juice')
print(s)

#Formatting strings using the format method - new style
s = 'Hello {} {}, it is a great {} to meet you at {}'.format('Mr.', 'Jones', 'pleasure', 5)
print(s)

s = 'Hello {} {}, it is a great {} to meet you at {} o\' clock'.format('Sir', 'Arthur', 'honor', 9)
print(s)

#Alternative ways of using string format
s = 'I have a {food_item} and a {drink_item} with me'.format(drink_item='soda', food_item='sandwich')
print(s)

s = 'The {animal} has the following attributes: {attributes}'.format(animal='dog', attributes=['lazy', 'loyal'])
print(s)

#--------------------------------Regular Expressions----------------------------#
s1 = 'Python is an excellent language'
s2 = 'I love the Python language. I also use Python to build applications at work!'

import re
pattern = 'python'
# match only returns a match if regex match is found at the beginning of the string
re.match(pattern, s1)
# pattern is in lower case hence ignore case flag helps
# in matching same pattern with different cases
re.match(pattern, s1, flags=re.IGNORECASE)

# printing matched string and its indices in the original string
m = re.match(pattern, s1, flags=re.IGNORECASE)
print('Found match {} ranging from index {} - {} in the string "{}"'.format(m.group(0),
m.start(),
m.end(), s1))

# match does not work when pattern is not there in the beginning of string s2
re.match(pattern, s2, re.IGNORECASE)

# illustrating find and search methods using the re module
re.search(pattern, s2, re.IGNORECASE)

re.findall(pattern, s2, re.IGNORECASE)

match_objs = re.finditer(pattern, s2, re.IGNORECASE)
match_objs

print("String:", s2)
for m in match_objs:
    print('Found match "{}" ranging from index {} - {}'.format(m.group(0),
                                                               m.start(), m.end()))

# illustrating pattern substitution using sub and subn methods
re.sub(pattern, 'Java', s2, flags=re.IGNORECASE)

re.subn(pattern, 'Java', s2, flags=re.IGNORECASE)

# dealing with unicode matching using regexes
s = u'H\u00e8llo! this is Python üêç'
s

re.findall(r'\w+', s)

re.findall(r"[A-Z]\w+", s, re.UNICODE)

emoji_pattern = r"['\U0001F300-\U0001F5FF'|'\U0001F600-\U0001F64F'|'\U0001F680-\U0001F6FF'|'\u2600-\u26FF\u2700-\u27BF']"
re.findall(emoji_pattern, s, re.UNICODE)

```
